Q.2 What deck were the passengers on and how does that relate to their class ?

```python
titanic_data_df.head()
```

```python
deck = titanic_data_df['Cabin'].dropna()
deck
```

Grab the first letter in cabin column to get the deck level

```python
levels = []
for level in deck:
    levels.append(level[0])
    
```

```python
cabin_df = pd.DataFrame(levels)
cabin_df.columns = ['Cabin']
cabin_df.sort_values(by = 'Cabin', inplace = True)
cabin_df
```

```python
cabin_df.Cabin.value_counts().plot(kind='bar', alpha=0.55)
plt.title("Passengers based on decks");
```

Q.3. Who was alone and who was with family

```python
titanic_data_df.head()
```

Add a new column to define Alone

```python
titanic_data_df['Alone'] = titanic_data_df.Parch + titanic_data_df.SibSp
```

```python
titanic_data_df
```

```python
titanic_data_df.loc[titanic_data_df['Alone'] > 0, 'Alone'] = 'with Family'
titanic_data_df.loc[titanic_data_df['Alone'] == 0, 'Alone'] = 'Alone'
```
