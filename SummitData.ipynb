```python
import pandas as pd
import numpy as np
datafile1 = your_local_path +'Rainier_Weather.csv'
datafile2 = your_local_path +'climbing_statistics.csv'
a = pd.read_csv(datafile1)
b = pd.read_csv(datafile2)
b = b.dropna(axis=1)
summit_data_df = a.merge(b, on='Date')
summit_data_df
```

```python
summit_data_df.describe()
```

```python
summit_data_df['Month_Val']=pd.to_datetime(summit_data_df['Date']).dt.month
```

```python
summit_data_df['Year']=pd.to_datetime(summit_data_df['Date']).dt.year
```

```python
# Based on year analyse the summit success rate 
Data_RainierByYear=summit_data_df.groupby(['Year'])["Attempted","Succeeded"].sum().reset_index()
```

```python
Data_RainierByYear['Success %']=(Data_RainierByYear['Succeeded']/Data_RainierByYear['Attempted'])*100
```

```python
# Removing redundant variables
summit_data_df_model=summit_data_df.drop(['Battery Voltage AVG','Wind Direction AVG', 'Solare Radiation AVG'], axis=1)
```

```python
summit_data_df_model.info()
```

```python
summit_data_df_model.isna().sum()
```

```python
#Split the data to train and test
input_cols = ['Temperature AVG', 'Relative Humidity AVG', 'Wind Speed Daily AVG', 'Attempted', 'Succeeded', 'Month_Val', 'Year']
y = summit_data_df_model['Success Percentage'].values
X = summit_data_df_model[input_cols]
```

```python
from sklearn import model_selection

test_size = 0.30 # taking 70:30 training and test set
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=test_size)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Creating Linear Regression model
linreg = LinearRegression()
linreg.fit(X_train, y_train)

y_predict = linreg.predict(X_test)
```

```python
# Root mean squared error (RMSE)
import numpy as np
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_predict, y_test)))
```

```python
linreg.score(X_test, y_test)
```
